```python
import pandas as pd

df = pd.read_csv("uci-news-aggregator.csv", sep=',', header=None, engine='python', on_bad_lines='skip')

df.columns = ['ID', 'Title', 'URL', 'Publisher', 'Category', 'Story', 'Hostname', 'Timestamp']

df.head()
```

```python
texts = df['Title']
texts.head()
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

tokens = texts.apply(word_tokenize)
tokens.head()
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    return [word for word in tokens if word.lower() not in stop_words]

filtered_tokens = tokens.apply(remove_stopwords)
filtered_tokens.head()
```

```python
clean_text = filtered_tokens.apply(lambda x: " ".join(x))
clean_text.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000)

X_tfidf = tfidf.fit_transform(clean_text)

X_tfidf.shape
```

```python
tfidf_df = pd.DataFrame(
    X_tfidf.toarray(),
    columns=tfidf.get_feature_names_out()
)

tfidf_df.head()
```

```python
from collections import Counter

all_words = " ".join(clean_text).split()
word_freq = Counter(all_words)

word_freq.most_common(10)
```
