```python
import pandas as pd

# Excel file ko read karna
df = pd.read_excel('/content/data/raw/apple_stock_10years_daywise.xlsx')

# Date ko datetime me convert karna
df['Date'] = pd.to_datetime(df['Date'])

# First few rows dekhna
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional styling
plt.style.use('seaborn-v0_8')
sns.set_theme()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['Date'], df['Close'], color='blue')
plt.title('Apple Stock Closing Price (2015â€“2025)')
plt.xlabel('Year')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.show()
```

Insight:
The line plot clearly shows a strong upward trend in Appleâ€™s stock price over the years. After 2020, the price increased significantly, indicating accelerated market growth and investor confidence. There are also small dips that reflect short-term market fluctuations.

```python
plt.figure(figsize=(8,5))
plt.hist(df['Close'], bins=30, color='green', edgecolor='black')
plt.title('Distribution of Closing Price')
plt.xlabel('Closing Price (USD)')
plt.ylabel('Frequency')
plt.show()
```

Insight:
The histogram reveals that most of the closing prices are concentrated between $50 and $150. This indicates that for a large portion of the decade, Appleâ€™s stock price remained in this stable range before rising sharply in later years.

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=df['High'], color='orange')
plt.title('Box Plot of High Prices')
plt.ylabel('Price (USD)')
plt.show()
```

Insight:
The box plot highlights several outliers in the high price data. These outliers represent days when Appleâ€™s stock price experienced unusual spikes, likely due to major product launches, market announcements, or broader economic events.

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Open'], df['Close'], alpha=0.5, color='purple')
plt.title('Open Price vs Close Price')
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.show()
```

Insight:
The scatter plot shows a clear positive correlation between opening and closing prices. When the opening price is high, the closing price also tends to be high. This indicates consistent daily trading behavior and stable market performance for Apple stock.

```python
df['Year'] = df['Date'].dt.year
yearly_avg = df.groupby('Year')['Close'].mean()

plt.figure(figsize=(12,6))
yearly_avg.plot(kind='bar', color='teal')
plt.title('Average Closing Price Per Year')
plt.xlabel('Year')
plt.ylabel('Average Close Price (USD)')
plt.xticks(rotation=45)
plt.show()
```

Insight:
The bar plot demonstrates a steady increase in the average closing price year over year. This upward trend is particularly sharp after 2020, reflecting Appleâ€™s strong financial performance, increased investor trust, and market dominance.
